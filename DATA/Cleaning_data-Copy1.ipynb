## Pseudoscorpions of the World
Harvey, M.S. (2013). Pseudoscorpions of the World, version 3.0. Western Australian Museum, Perth. http://www.museum.wa.gov.au/catalogues/pseudoscorpions

```python
references = "PBDB_data/references2.csv"
citation_data = pd.read_csv(references)

citation_data
```

```python
# Import our data into pandas from CSV
author_string = 'PBDB_data/references2.csv'
author_df = pd.read_csv(author_string)

author_df
```

```python
# Create a group based on the values in the 'author1last' column
author_group = author_df.groupby('author1last')

# Count how many times each author appears in our group
count_authors = author_group['author1last'].count()

count_authors
```

```python
# Create a bar chart based off of the group series from before
author_chart = count_authors.plot(kind='bar')

# Set the xlabel and ylabel using class methods
author_chart.set_xlabel("Author")
author_chart.set_ylabel("Number of Works Cited")

plt.tight_layout()
plt.savefig("NumWorksCited.png")
plt.show()
```

```python
# Create a group based on the values in the 'pubyr' column
pubyr_group = author_df.groupby('pubyr')

# Count how many times each pubyr appears in our group
count_pubyr = pubyr_group['pubyr'].count()

count_pubyr
```

```python
# Create a bar chart based off of the group series from before
author_chart = count_pubyr.plot(kind='line')

# Set the xlabel and ylabel using class methods
author_chart.set_xlabel("Publish Year")
author_chart.set_ylabel("Number of Authors Published")

plt.tight_layout()
plt.savefig("NumAuthPub.png")
plt.show()
```
